```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "/Users/amirgazar/Documents/GitHub/Decarbonization-Tradeoffs/1 Decarbonization Pathways/Decarbonization_Pathways_Land_Use.xlsx"

# Read all sheets into a dictionary
all_sheets = pd.read_excel(file_path, sheet_name=None)

# List to hold DataFrames
df_list = []

# Iterate over each sheet
for sheet_name, df in all_sheets.items():
    df['Pathway'] = sheet_name  # Add the 'Pathway' column
    df_list.append(df)

# Concatenate all DataFrames into one
decarb_pathways = pd.concat(df_list, ignore_index=True)
```

```python
decarb_pathways
```

```python
# 1. grab all the "(ha)" columns
ha_cols = [c for c in decarb_pathways.columns if '(ha)' in c]

# 2. make a copy (or you can do it in‐place) and convert to 10 000 km²
df = decarb_pathways.copy()
df[ha_cols] = df[ha_cols] / 1e6

# 3. group by Pathway and sum across years
sum_by_pathway = (
    df
    .groupby('Pathway')[ha_cols]
    .sum()
    .reset_index()
)

sum_by_pathway
```

```python
import matplotlib.pyplot as plt

# assume sum_by_pathway is already your DataFrame, with 
# a “Pathway” column plus one column per land-use tech

# 1. put “Pathway” into the index
df_plot = sum_by_pathway.set_index('Pathway')

# 2. plot a stacked bar chart
ax = df_plot.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    legend=True
)

ax.set_xlabel('Pathway')
ax.set_ylabel('Total area (in units of 10 000 km²)')
ax.set_title('Land-use breakdown by pathway')
plt.xticks(rotation=0)      # pathway labels horizontal
plt.tight_layout()
plt.show()
```

```python
import re
from math       import ceil
import numpy     as np
import matplotlib.pyplot as plt
import matplotlib.patches  as mpatches
import matplotlib.colors   as mcolors

# your custom palette
tech_palette = {
    "Old fossil":    "#D9BCA9",
    "New NG":        "#E9967A",
    "Nuclear":       "#C27BA0",
    "SMR":           "#EFD9B4",
    "Hydropower":    "#9EB1D4",
    "Biomass":       "#A2D9B1",
    "Imports QC":    "#D1B1D6",
    "Imports NYISO": "#F4B6C2",
    "Imports NBSO":  "#F2A2B0",
    "Solar":         "#FBE7A1",
    "Onshore Wind":  "#B4D6E3",
    "Offshore Wind": "#A9CAD6"
}

# 0. Copy & normalize columns
df = decarb_pathways.copy()
df.columns = df.columns.str.strip()

# 1. Find all land‐use columns
ha_cols = [c for c in df.columns if c.endswith('(ha)')]

# 2. Convert to “blocks” of 10 000 km²
df_blocks = df.copy()
df_blocks[ha_cols] = df_blocks[ha_cols] / 1e6

# 3. Sum up by pathway
land_units = df_blocks.groupby('Pathway')[ha_cols].sum()

# 4. Choose resolution so the smallest non-zero chunk → ≥1 cube
min_nonzero = land_units[land_units>0].stack().min()
resolution  = ceil(1 / min_nonzero)
print(f"Using resolution = {resolution} sub-blocks per 10 000 km²")

# 5. Compute integer cube counts
sub_blocks = (land_units * resolution).round().astype(int)

# 6. Map each ha-column to your tech key
tech_keys = {}
for ha in ha_cols:
    base = re.sub(r'\s+land use \(ha\)$', '', ha)
    if base.startswith('Onshore Wind'):
        base = 'Onshore Wind'
    if base.lower().startswith('canadian hydro'):
        base = 'Hydropower'
    tech_keys[ha] = base

# 7. Assign colors
ha_colors = {ha: tech_palette[tech_keys[ha]] for ha in ha_cols}

# 8. Build the figure
pathways = sub_blocks.index.tolist()
n = len(pathways)
cols = 4
rows = ceil(n/cols)
fig = plt.figure(figsize=(4*cols, 4*rows), constrained_layout=True)

for idx, pw in enumerate(pathways):
    counts = sub_blocks.loc[pw]
    total  = counts.sum()
    if total == 0:
        continue

    # pack into an L×L×L grid with gaps (size = 2L-1)
    L    = ceil(total**(1/3))
    size = 2*L - 1
    mask = np.zeros((size, size, size), dtype=bool)
    facecolors = np.zeros((size, size, size, 4), dtype=float)

    # positions at even indices
    pos_list = [(2*i, 2*j, 2*k)
                for i in range(L)
                for j in range(L)
                for k in range(L)]

    # build flat list of ha keys
    flat_tech = []
    for ha in ha_cols:
        flat_tech += [ha] * counts[ha]

    # place cubes
    for (x, y, z), ha in zip(pos_list, flat_tech):
        mask[x, y, z] = True
        facecolors[x, y, z] = mcolors.to_rgba(ha_colors[ha])

    ax = fig.add_subplot(rows, cols, idx+1, projection='3d')
    ax.voxels(mask,
              facecolors=facecolors,
              edgecolor='gray',
              linewidth=0.3)
    ax.set_box_aspect((size, size, size))
    ax.view_init(elev=30, azim=45)
    ax.set_axis_off()
    ax.set_title(f'Pathway {pw}', fontsize=10)

# 9. Unified legend
handles = []
seen = set()
for ha in ha_cols:
    tech = tech_keys[ha]
    if tech not in seen:
        seen.add(tech)
        handles.append(mpatches.Patch(color=tech_palette[tech], label=tech))

fig.legend(handles,
           [h.get_label() for h in handles],
           loc='right',
           title='Technology',
           frameon=False,
           fontsize=9,
           title_fontsize=10)

# 10. Save and show
fig.savefig('land_use_floating_cubes.svg', format='svg')
plt.show()
```
