```python
import pandas as pd

file_path = "/Users/amirgazar/Documents/GitHub/EPA_Debarbonization/4 External Data/Survey Results/survey_results.xlsx"
survey_data = pd.read_excel(file_path)  
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Assuming 'survey_data' is your DataFrame
# Replace this line with actual data loading if necessary
# survey_data = pd.read_excel("path_to_your_file.xlsx")

# Combine all responses in the Q10 column
q10_responses = survey_data['Q10'].dropna().astype(str)

# Tokenize words (split by whitespace, and convert to lowercase)
all_words = ' '.join(q10_responses).lower().split()

# Enhanced stopwords list
stopwords = set([
    "and", "the", "of", "to", "in", "a", "on", "for", "with", "is", "that", "by", 
    "at", "an", "or", "i", "we", "our", "you", "it", "be", "are", "from", "was", 
    "not", "as", "but", "have", "this", "will", "can", "my", "no", "all", "if", 
    "they", "your", "which", "has", "been", "us", "more", "would", "any", "so",
    "should", "had", "these", "use", ".", "india", "their", "than", "about", "like", "having"
])
filtered_words = [word for word in all_words if word not in stopwords]

# Count word frequencies
word_counts = Counter(filtered_words)

# Get the most common words
common_words = word_counts.most_common(10)

# Plot a bar chart of the most common words
plt.figure(figsize=(10, 6))
plt.bar([word for word, count in common_words], [count for word, count in common_words])
plt.title("Most Common Specific Words in Q10 Responses")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()

# Optional: Create a Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
#plt.title("Word Cloud of Specific Words in Q10 Responses")
# Save the word cloud as a vector graphic (SVG format)
output_path = "Documents/GitHub/EPA_Debarbonization/5 Figures/Survey Results Analysis/survey_results_wordcloud.svg"
plt.savefig(output_path, format='svg', bbox_inches='tight')
plt.show()
```
